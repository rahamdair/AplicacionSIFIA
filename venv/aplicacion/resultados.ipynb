```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos Desbalanceados - Análisis de resultados
data = {
    "Algoritmo": ["RF", "XGBoost", "MLPNN", "SVM", "AdaBoost", "DT", "RL", "DNN", "CNN", "LSTM"],
    "Accuracy": [0.87, 0.87, 0.87, 0.86, 0.86, 0.84, 0.82, 0.86, 0.85, 0.85],
    "Recall": [0.46, 0.46, 0.49, 0.37, 0.45, 0.43, 0.17, 0.39, 0.53, 0.30],
    "Precision": [0.79, 0.80, 0.74, 0.83, 0.73, 0.63, 0.60, 0.80, 0.74, 0.82],
    "F1-Score": [0.58, 0.59, 0.59, 0.51, 0.56, 0.51, 0.27, 0.53, 0.58, 0.44]
}

# Crear el DataFrame de los resultados con los datos desbalanceados
df_resultados = pd.DataFrame(data)

# Configuración de gráficos agrupados
def plot_grouped_bars(df_resultados, metrics, title, xlabel, ylabel, colors):
    x = np.arange(len(df_resultados['Algoritmo']))  # posiciones en el eje X
    width = 0.35  # ancho de las barras

    fig, ax = plt.subplots(figsize=(10, 6))

    # Gráficas agrupadas
    bars1 = ax.bar(x - width / 2, df_resultados[metrics[0]], width, label=metrics[0], color=colors[0])
    bars2 = ax.bar(x + width / 2, df_resultados[metrics[1]], width, label=metrics[1], color=colors[1])

    # Etiquetas y títulos
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(df_resultados['Algoritmo'])
    ax.legend()
    ax.set_ylim(0, 1.1)  # Definir límites del eje Y

    # Añadir valores encima de las barras
    for bars in [bars1, bars2]:
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height:.2f}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),  # desplazamiento
                        textcoords="offset points",
                        ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
```

```python
# Gráfico 1: Accuracy y Recall
plot_grouped_bars(
    df_resultados,
    metrics=["Accuracy", "Recall"],
    title="Accuracy y Recall por Algoritmo",
    xlabel="Algoritmos de ML y DL",
    ylabel="Valores",
    colors=["#1f77b4", "#ff7f0e"]
)

# Gráfico 2: Precision y F1-Score
plot_grouped_bars(
    df_resultados,
    metrics=["Precision", "F1-Score"],
    title="Precisión y F1-Score por Algoritmo",
    xlabel="Algoritmos de ML y DL",
    ylabel="Valores",
    colors=["#1f77b4", "#ff7f0e"]
    #colors=["#2ca02c", "#d62728"]
)
```
